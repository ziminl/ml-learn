### 2.2. Изменение размера графика

Параметр figsize, передаваемый функции plt.subplots() позволяет изменить размер графика.

Параметр принимает на вход кортеж (tuple) из двух чисел:
   
1 - Ширина (дюймы)

2 - Высота (дюймы)

```python
import matplotlib.pyplot as plt
% matplotlib inline

fig, ax = plt.subplots(figsize=(10,5))
```

figsize работает совместно с заданием числа графиков в окне. Однако следует учитывать, что figsize задаёт график всего окна, а не каждого отдельного графика.

```python
import matplotlib.pyplot as plt
% matplotlib inline

fig, ax = plt.subplots(3,1, figsize=(10,5))
```

### 2.3. Построение графика.

#### 2.3.1. Способы задания графика.

##### 2.3.1.1. List

2 мерный график можно по точкам, используя списки координат. Проиллюстрируем на 1 графике:

```python
import matplotlib.pyplot as plt
% matplotlib inline

fig, ax = plt.subplots()

x = [1, 2, 3, 4, 5] 
y = [0, 2, 9, -1, 13]
ax.plot(x, y)
plt.show()
```

##### 2.3.1.2. Numpy array

MatPlotlib принимает значения в виде Numpy array. Для этого сгенерируем массив аргументов из 10 на интервале от -2 до 2. А массив функций от аргументов получим из них, возведя каждый элемент в квадрат:

```python
import matplotlib.pyplot as plt
% matplotlib inline

import numpy # не забудем импортировать Numpy!

fig, ax = plt.subplots()

x = numpy.linspace(-2,2,10) # Сгенерируем Numpy массив из 10 точек (аргументов, x) в интервале от -2 до 2.
y = x**2

ax.plot(x, y) # Построим график f(x)=x**2
plt.show() # Покажем результат построений
```

##### 2.3.1.3. Задание функции сразу в теле ax.plot()

Не обязательно получать значения заранее. Вычислять значение функции можно сразу внутри построения графика:

```python
import matplotlib.pyplot as plt
% matplotlib inline

import numpy # не забудем импортировать Numpy!

fig, ax = plt.subplots()

x = numpy.linspace(-2,2,5) # Сгенерируем Numpy массив из 5 точек (аргументов, x) в интервале от -2 до 2.

ax.plot(x, x**3) # Построим график f(x)=x**3
plt.show() # Покажем результат построений
```

##### 2.3.1.4. Задание 2 графиков на 1 поле

Графики можно строить как последовательным вызовом ax.plot():

```python
import matplotlib.pyplot as plt
% matplotlib inline

import numpy # не забудем импортировать Numpy!

fig, ax = plt.subplots()

x = numpy.linspace(-2,2,10) # Сгенерируем Numpy массив из 10 точек (аргументов, x) в интервале от -2 до 2.

ax.plot(x, x**2) # Построим график f(x)=x**2
ax.plot(x, x**3) # Построим график f(x)=x**3
plt.show() # Покажем результат построений
```

так и вызвав аргументы и функции от аргумента внутри 1 ax.plot() последовательными парами:

$ax.plot(x_{1}, f(x_{1}), x_{2}, f(x_{2}), ... , x_{n}, f(x_{n}))$

```python
import matplotlib.pyplot as plt
% matplotlib inline

import numpy # не забудем импортировать Numpy!

fig, ax = plt.subplots()

x = numpy.linspace(-1,1,9) # Сгенерируем Numpy массив из 9 точек (аргументов, x) в интервале от -1 до 1.

ax.plot(x, x**2, x, x**3, x, x**4) # построим последовательно 3 графика внутри 1 функции plot()
plt.show() # Покажем результат построений
```

##### 2.3.1.5. Задание нескольких графиков в разных осях и/или окнах

* Для построения графика в отдельном блоке осей внутри одного окна достаточно указать индекс окна у объекта ax при построении графика
* Для построения графика в отдельном окне необходимо указать объект осей, соответвующий этому окну.

Для примера построим 3 графика:
* 1 в 1 окне в 1 наборе осей
* по 1 в каждом из 2 набре осей, во 2 окне

```python
import matplotlib.pyplot as plt
% matplotlib inline

import numpy

x = numpy.linspace(-1,1,9) # Сгенерируем Numpy массив из 9 точек (аргументов, x) в интервале от -1 до 1.

fig, ax = plt.subplots(1,1) # 1-е окно, 1 график в 1 строке
fig2, ax2 = plt.subplots(1,2) # 2-е окно, 2 графика в 1 строке

ax.plot(x, x**3)
ax2[0].plot(x, x**2)
ax2[1].plot(x, x**4)
```
